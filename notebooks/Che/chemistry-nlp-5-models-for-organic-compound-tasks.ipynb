<font size="+2" color='#FF8C00'><b> ✨ DistilBERT for Question Answering:</b></font>

```python
from transformers import DistilBertTokenizer, DistilBertForQuestionAnswering, pipeline

# Load DistilBERT model and tokenizer
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-cased-distilled-squad')
model = DistilBertForQuestionAnswering.from_pretrained('distilbert-base-cased-distilled-squad')


# Example usage
question_answering = pipeline('question-answering', model=model, tokenizer=tokenizer)
result = question_answering(question='What is the IUPAC name for the organic compound?', context='There can be several isomers with the molecular formula C4H8O.')
print(result)
```

```python
# Example 1
result = question_answering(question='What is the systematic name of the organic compound?', context='The systematic name of the organic compound will be provided.')
print(result)

# Example 2
result = question_answering(question='Provide the IUPAC name of the following compound.', context='The systematic name of the compound CH3CH2CH2O will be provided.')
print(result)

# Example 3
result = question_answering(question='What is the IUPAC name of the given organic compound?', context='The IUPAC name of the given organic compound (with structure) will be provided.')
print(result)
```

<font size="+2" color='#FF8C00'><b> ✨ Text Summarization using Latent Semantic Analysis (LSA)</b></font>


<div style="border-radius: 10px; border: 2px solid #4CAF50; padding: 15px; background-color: #D1F2A5; font-size: 100%; text-align: left;">
    
<font size="+1" color="#059c99"><b>Text Summarization with LSA:</b></font>

**1. Input Text:**

* The process begins with a given input text that needs to be summarized. This text can be any document, article, or piece of content.

**2. Parsing the Text:**

* The raw text is parsed to extract meaningful information. During parsing, the text is tokenized into sentences and words, preparing it for further analysis.

**3. Latent Semantic Analysis (LSA):**

* LSA is a mathematical technique used in natural language processing and information retrieval. It analyzes the relationships between words in a given context and identifies latent (hidden) semantic structures.

**4. Vector Representation:**

* LSA represents the words and sentences in the text as vectors in a high-dimensional space. These vectors capture the semantic meaning of the words based on their co-occurrence patterns.

**5. Dimensionality Reduction:**

* LSA involves reducing the dimensionality of the vector space to identify the most important dimensions (semantic concepts). This step helps in capturing the core semantic relationships while discarding noise.

**6. Sentence Ranking:**

* After dimensionality reduction, sentences are ranked based on their importance in capturing the underlying semantic structure of the text. Sentences that contribute most to the overall meaning are given higher ranks.

**7. Summary Generation:**

* The highest-ranked sentences are selected to form a concise summary of the input text. The number of sentences selected can be adjusted based on the desired length of the summary.

**8. Output Summary:**

* The final output is a summary that retains the key information and semantic structure of the original text. This summary is generated through the identification of significant sentences using the LSA algorithm.

**Advantages of LSA-based Summarization:**

* LSA considers the semantic meaning of words, providing a more contextually aware summarization.
* It captures latent relationships in the text, allowing for a more nuanced understanding of content.
* LSA-based summarization is particularly useful when dealing with large datasets and documents.

**Considerations:**

* The quality of the summary depends on the effectiveness of the LSA algorithm and the inherent semantic structure of the text.
* It's an extractive summarization method, meaning it selects and combines existing sentences rather than generating entirely new sentences.


```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer

# Example usage
text = "Isomerism is a phenomenon in which two or more organic compounds have the same molecular formula but different arrangements of atoms."

# Parse the text
parser = PlaintextParser.from_string(text, Tokenizer("english"))

# Initialize the LsaSummarizer
summarizer = LsaSummarizer()

# Get the summary
summary = summarizer(parser.document, 1)  # You can adjust the number of sentences in the summary

# Print the result
for sentence in summary:
    print(sentence)
```

<font size="+2" color='#FF8C00'><b> ✨ T5 Transformer for Abstractive Summarization:</b></font>



```python
from transformers import T5ForConditionalGeneration, T5Tokenizer

# Example usage
tokenizer = T5Tokenizer.from_pretrained("t5-small")
model = T5ForConditionalGeneration.from_pretrained("t5-small")
text = "Isomerism is a phenomenon in which two or more organic compounds have the same molecular formula but different arrangements of atoms."
inputs = tokenizer("summarize: " + text, return_tensors="pt", max_length=512, truncation=True)
summary_ids = model.generate(inputs["input_ids"])
result = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
print(result)
```

<font size="+2" color='#FF8C00'><b> ✨ BART (Bidirectional and Auto-Regressive Transformers) model</b></font>


```python
from transformers import BartForConditionalGeneration, BartTokenizer

# Example usage
tokenizer = BartTokenizer.from_pretrained("facebook/bart-large-cnn")
model = BartForConditionalGeneration.from_pretrained("facebook/bart-large-cnn")

text = "Isomerism is a phenomenon in which two or more organic compounds have the same molecular formula but different arrangements of atoms."
inputs = tokenizer.encode("summarize: " + text, return_tensors="pt", max_length=1024, truncation=True)

summary_ids = model.generate(inputs, max_length=150, min_length=50, length_penalty=2.0, num_beams=4, early_stopping=True)
result = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
print(result)
```
